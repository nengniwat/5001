```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("https://github.com/nengniwat/5001/blob/main/English%20proficiency%20data.csv?raw=True")
df = df[:54]
df
```

```python
df1 = pd.DataFrame(df.groupby('Sex')[['BUU_CET_Score','Listening','Reading']].mean())
df1
```

```python
fig = px.bar(df1, x=df1.index, y=df1.columns, barmode='group')

fig.update_xaxes(title='English Score')
fig.show()
```
